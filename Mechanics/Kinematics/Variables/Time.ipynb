# Solve for Time ($t$): 
# $$t = \frac{x_f - x_i}{v}$$
This equation answers the question: How long does it take to cross a specific gap in space at a specific rate of speed? Instead of distributing space over time to find a rate, you are doing the reverse. You take the total gap of space you need to conquer ($x_f - x_i$) and divide it by your velocity ($v$). In an absolute sense, division here means grouping. You are counting how many times your velocity package (e.g., chunks of 3 meters) can fit into the total distance package.
