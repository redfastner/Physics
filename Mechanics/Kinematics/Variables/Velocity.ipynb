# Solve Average Velocity  
By definition, velocity ($v$) is the rate at which an object's position changes over a specific interval of time. Mathematically, it is written as:  
# $$v =  \frac{\Delta x}{\Delta t}$$ 
Syntax: v = \frac{\Delta x}{\Delta t}

$\Delta x$ means "change in position" (Displacement) = $x_f - x_i$  

$\Delta t$ means "change in time" (Time elapsed) = $t_f - t_i$  

If we start our stopwatch at zero ($t_i = 0$), then the time elapsed is just the final time reading, which we call $t$. So, $\Delta t = t$.

### Breaking Down the Changes ($\Delta$)

Let's substitute these two expansions back into our starting definition:

$$v = \frac{\text{(displcement) } x_f - x_i}{t}$$

Why Subtraction? ($x_f - x_i$)  

Subtraction is the mathematical operation for finding the **difference** or the "gap" between two states.  

If you added them ($17 + 5 = 22$), the number 22 tells you nothing about the path traveled; it scales up based on how far away you are from some arbitrary zero point.  

If you multiplied them ($17 \times 5 = 85$), the result scales non-linearly.  

Why Time? ($t$)  

Time is our universal metric for *rate*. In the absolute physical sense, time is the standard cosmic backdrop that flows forward at a constant, unceasing pace. Because it flows independently of the object, it acts as the perfect neutral yardstick to measure how "fast" or "slow" the change in space occurred.

Why not something else, like fuel used or heartbeats? Because those variables change depending on the object. Time applies equally to a rock, a planet, or a runner. It is the fundamental baseline of the universe.  

Why Division? ($\frac{\text{Space}}{\text{Time}}$)  

We use division because velocity is fundamentally a **rate per single unit of time** (e.g., meters *per one single second*).  

The flaw of multiplication: If you multiplied space and time ($12 \text{ meters} \times 4 \text{ seconds} = 48$), the number gets *bigger* the longer you travel. That implies you are moving "more" or "faster" just because you let the clock run longer. That defies the concept of a steady speed.  

The meaning of division:** Division means "distribution." When you calculate $\frac{12 \text{ meters}}{4 \text{ seconds}}$, you are taking the total chunk of space you conquered ($12\text{m}$) and chopping it up equally across every single second that ticked by.

# Instantaneous Velocity
# $$v(t) = \lim_{\Delta t \to 0} \frac{x(t + \Delta t) - x(t)}{\Delta t}$$

Syntax: v(t) = \lim_{\Delta t \to 0} \frac{x(t + \Delta t) - x(t)}{\Delta t}  

Where:  

$v(t)$ means velocity as a function of time, plug in a number into t and the number appears in the equation for t, this is what we are technically solving for.  

Lim $\Delta{t \to 0}$ means time elapsed (it is always tf - ti), as it approaches 0 (lim, eg 0.44, 0.22, 0.2043, $\Delta t = 0.000000001\text{ seconds}$, till 0)
