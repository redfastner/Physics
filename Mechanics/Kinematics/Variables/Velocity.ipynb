# Average Velocity  
By definition, velocity ($v$) is the rate at which an object's position changes over a specific interval of time. Mathematically, it is written as:  
# $$v =  \frac{\Delta x}{\Delta t}$$ 
Syntax: v = \frac{\Delta x}{\Delta t}

$\Delta x$ means "change in position" (Displacement) = $x_f - x_i$  

$\Delta t$ means "change in time" (Time elapsed) = $t_f - t_i$  

If we start our stopwatch at zero ($t_i = 0$), then the time elapsed is just the final time reading, which we call $t$. So, $\Delta t = t$.

### Breaking Down the Changes ($\Delta$)

Let's substitute these two expansions back into our starting definition:

$$v = \frac{\text{(displcement) } x_f - x_i}{t}$$

Why Subtraction? ($x_f - x_i$)  

Subtraction is the mathematical operation for finding the **difference** or the "gap" between two states.  

If you added them ($17 + 5 = 22$), the number 22 tells you nothing about the path traveled; it scales up based on how far away you are from some arbitrary zero point.  

If you multiplied them ($17 \times 5 = 85$), the result scales non-linearly.  

Why Time? ($t$)  

Time is our universal metric for *rate*. In the absolute physical sense, time is the standard cosmic backdrop that flows forward at a constant, unceasing pace. Because it flows independently of the object, it acts as the perfect neutral yardstick to measure how "fast" or "slow" the change in space occurred.

Why not something else, like fuel used or heartbeats? Because those variables change depending on the object. Time applies equally to a rock, a planet, or a runner. It is the fundamental baseline of the universe.  

Why Division? ($\frac{\text{Space}}{\text{Time}}$)  

We use division because velocity is fundamentally a **rate per single unit of time** (e.g., meters *per one single second*).  

The flaw of multiplication: If you multiplied space and time ($12 \text{ meters} \times 4 \text{ seconds} = 48$), the number gets *bigger* the longer you travel. That implies you are moving "more" or "faster" just because you let the clock run longer. That defies the concept of a steady speed.  

The meaning of division:** Division means "distribution." When you calculate $\frac{12 \text{ meters}}{4 \text{ seconds}}$, you are taking the total chunk of space you conquered ($12\text{m}$) and chopping it up equally across every single second that ticked by.

# Instantaneous Velocity
# $$v(t) = \lim_{\Delta t \to 0} \frac{x(t + \Delta t) - x(t)}{\Delta t}$$
Syntax: v(t) = \lim_{\Delta t \to 0} \frac{x(t + \Delta t) - x(t)}{\Delta t}  

Where:  

$v(t)$: velocity as a function of time, plug in a number into t and the number appears in the equation for t, this is what we are solving for.  

Lim $\Delta{t \to 0}$: time elapsed (it is always tf - ti), as it approaches 0 (lim, eg 0.44, 0.22, 0.2043, $\Delta t = 0.000000001\text{ seconds}$, till 0).  

$x(t + \Delta t)$: x is not a variable multiplying t, x is the function name "Position", the parentheses mean "at the specific clock reading of" so the expression translates to - the coordinate position of the object on the map, evaluated at the exact clock reading of your initial time (t) plus a tiny sliver of elapsed time i.e it is a bulletproof way of writing final position. If you remove the $x$, you are left with just $(t + \Delta t)$ which is just a clock reading (seconds). We aren't trying to calculate time; we are trying to calculate where the object is in space. The $x$ must stay because it takes that clock reading and maps it to a physical location in meters. Without the $x$, the physics loses its connection to space. If you remove the $t$, you might try to write something like $x(\Delta t)$. $x(\Delta t)$ would mean "your position after a duration of time has passed, assuming you started at zero." But what if your stopwatch didn't start at zero? What if you started tracking the object at $t = 5\text{ seconds}$? The expression requires the base $t$ because it pins your starting moment to a specific location on the cosmic timeline. Therefore, $(t + \Delta t)$ represents the direction of time. $t$ is your present moment anchor. $+$ means time is moving forward. $\Delta t$ is the future step you are taking.

$- x(t)$: $x$ is the machine that tracks position (meters), $(t)$ is the input slot where you drop a clock reading (seconds), i.e $x_i$ (Position evaluated at the current/initial time). Why must it be subtraction ($x_f - x_i$), and not addition or multiplication? Addition ($x_f + x_i$) fails because if you started at coordinate 10m and moved to coordinate 12m, adding them gives 22. The number 22 scales up purely based on how far away you are from some arbitrary zero-point on the track. It tells you nothing about the actual spatial gap you crossed. Multiplication ($x_f \times x_i$) fails because $10 \times 12 = 120$. It scales non-linearly, blowing the number up exponentially. Subtraction ($x_f - x_i$) is the only operation that isolates the gap. $12\text{m} - 10\text{m} = 2\text{m}$. Subtraction actively strips away your starting anchor ($x_i$), leaving you with the pure, raw change in space. We call this displacement ($\Delta x$).

Divide by the Change in Time ($\Delta t$): Division means uniform distribution. When you shrink the stopwatch window down to an infinitesimally small flash of time ($\Delta t \to 0$), your object covers an infinitesimally small slice of space ($\Delta x$). Let's say that in a tiny flash lasting $0.000001\text{ seconds}$, an object creeps forward by $0.000003\text{ meters}$. If you just looked at the raw distance ($0.000003\text{ meters}$), it seems utterly insignificant. But when you divide by that tiny sliver of time, you are performing an act of projection: $$\frac{0.000003\text{ meters}}{0.000001\text{ seconds}} = 3\text{ m/s}$$ By dividing, you take that tiny, microscopic behavior and ask: "If the universe kept this motion perfectly steady for one full, massive second, how much total space would be conquered?" Division normalizes the flash. It scales the micro-movement up into a standard, universal metric: meters per single unit of time.

What Exactly is $v(t)$ when Solved?  

A customized blueprint equation that reveals the exact intensity and exact direction of an object's motion at any frozen millisecond you choose on the timeline. You drop in $t = 1\text{s}$, the machine spits out $+3\text{ m/s}$. (Moving forward at a rate of 3 meters per second right now). You drop in $t = 4\text{s}$, the machine spits out $+48\text{ m/s}$. (Moving forward at a blistering rate of 48 meters per second right now).

The entire fraction: $$v(t) = \lim_{\Delta t \to 0} \frac{x(t + \Delta t) - x(t)}{\Delta t}$$  

Translated: $$\text{Numerator: } \frac{\text{Output at Future Time } (x_f) - \text{Output at Present Time } (x_i)}{\text{Denominator: } \text{Input Future Time } (t_f) - \text{Input Present Time } (t_i)}$$
