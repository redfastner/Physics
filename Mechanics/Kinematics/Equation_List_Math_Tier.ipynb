## 1. Constant Velocity (The Ground Floor)

Use these only when acceleration is exactly zero ($a = 0$).

* **Solve for Position ($x_f$):** 
$$x_f = x_i + vt$$


* **Solve for Constant/Average Velocity ($v$):** 
$$v = \frac{x_f - x_i}{t}$$


* **Solve for Time ($t$):** 
$$t = \frac{x_f - x_i}{v}$$

---

## 2. Constant Acceleration: The Big 4 (The Core Toolkit)

Use these when speed is changing, but changing at a constant, steady rate.

### Base Equations (As usually memorized)

* $$v = v_0 + at$$


* $$\Delta x = v_0t + \frac{1}{2}at^2$$


* $$v^2 = v_0^2 + 2a\Delta x$$


* $$\Delta x = \left(\frac{v_0 + v}{2}\right)t$$



### The "Solve For" Master Variations

* **Solve for Position ($x_f$):**

$$x_f = x_i + v_0t + \frac{1}{2}at^2$$


$$x_f = x_i + \frac{v^2 - v_0^2}{2a}$$


* **Solve for Initial Velocity ($v_0$):**

$$v_0 = v - at$$


$$v_0 = \frac{\Delta x}{t} - \frac{1}{2}at$$


$$v_0 = \sqrt{v^2 - 2a\Delta x}$$


* **Solve for Final Velocity ($v$):**

$$v = v_0 + at$$


$$v = \sqrt{v_0^2 + 2a\Delta x}$$


* **Solve for Acceleration ($a$):**

$$a = \frac{v - v_0}{t}$$


$$a = \frac{2(\Delta x - v_0t)}{t^2}$$


$$a = \frac{v^2 - v_0^2}{2\Delta x}$$


* **Solve for Time ($t$):**

$$t = \frac{v - v_0}{a}$$


$$\frac{1}{2}at^2 + v_0t - \Delta x = 0 \quad \rightarrow \text{ (Solve for } t \text{ using the Quadratic Formula Setup)}$$

## 3. Two-Dimensional Motion (The X & Y Split)

For projectile motion, you duplicate the tools above into independent perpendicular dimensions.

### Horizontal Axis ($a_x = 0$)

* $$x_f = x_i + v_{0x}t$$


* $$v_x = v_{0x} = v_0\cos\theta$$



### Vertical Axis ($a_y = -g$)

* $$y_f = y_i + v_{0y}t - \frac{1}{2}gt^2$$


* $$v_y = v_{0y} - gt$$


* $$v_y^2 = v_{0y}^2 - 2g\Delta y$$


* $$v_{0y} = v_0\sin\theta$$



---

## 4. Calculus Kinematics (The Final Boss)

Use these when acceleration is **not** constant (it changes over time, like $a(t) = 3t^2$). The standard algebraic formulas above become illegal here.

### Going Down the Ladder (Derivatives / Slopes)

* **Solve for Velocity Function ($v(t)$):**

$$v(t) = \frac{dx}{dt}$$


* **Solve for Acceleration Function ($a(t)$):**

$$a(t) = \frac{dv}{dt} = \frac{d^2x}{dt^2}$$



### Going Up the Ladder (Integrals / Area Under Curve)

* **Solve for Velocity Function ($v(t)$):**

$$v(t) = v_0 + \int_{0}^{t} a(t) \, dt$$


* **Solve for Position Function ($x(t)$):**

$$x(t) = x_i + \int_{0}^{t} v(t) \, dt$$
