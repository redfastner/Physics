# Phase 1: One-Dimensional (1D) Kinematics

## 1. Constant Velocity & Core Definitions (Syllabus Sections 1–6)

*Use these when acceleration is zero ($a=0$). Direction matters (velocity is a vector with +/- signs), but path length (distance/speed) ignores direction.*

* **Displacement ($\Delta x$):**

$$\Delta x = x_f - x_i$$


* **Solve for Position ($x_f$):** 
$$x_f = x_i + vt$$


* **Solve for Constant / Average Velocity ($v_{\text{avg}}$):** 
$$v_{\text{avg}} = \frac{x_f - x_i}{t}$$


* **Solve for Time ($t$):** 
$$t = \frac{x_f - x_i}{v}$$

## 2. Instantaneous Velocity & Basic Calculus (Syllabus Section 3)

*Rice introduces this early as the "slope at an exact moment" on a kinematics graph. Use this when the window of time shrinks to zero.*

* **Instantaneous Velocity ($v(t)$):**

$$v(t) = \frac{dx}{dt}$$



---

## 3. Constant Acceleration: The Big 4 Toolkit (Syllabus Sections 7–8)

*Use these when speed is changing steadily. Rice focuses on standard algebra derivations here, heavily featuring gravity ($a = -g = -9.8 \text{ m/s}^2$) for vertical "Chalk Throwing" style problems.*

### Master Variations

* **Solve for Position ($x_f$):**

$$x_f = x_i + v_0t + \frac{1}{2}at^2$$


$$x_f = x_i + \frac{v^2 - v_0^2}{2a}$$


* **Solve for Initial Velocity ($v_0$):**

$$v_0 = v - at \quad \mid \quad v_0 = \frac{\Delta x}{t} - \frac{1}{2}at \quad \mid \quad v_0 = \sqrt{v^2 - 2a\Delta x}$$


* **Solve for Final Velocity ($v$):**

$$v = v_0 + at \quad \mid \quad v = \sqrt{v_0^2 + 2a\Delta x}$$


* **Solve for Acceleration ($a$):**

$$a = \frac{v - v_0}{t} \quad \mid \quad a = \frac{2(\Delta x - v_0t)}{t^2} \quad \mid \quad a = \frac{v^2 - v_0^2}{2\Delta x}$$


* **Solve for Time ($t$):**

$$t = \frac{v - v_0}{a} \quad \rightarrow \quad \text{Or use Quadratic Setup: } \frac{1}{2}at^2 + v_0t - \Delta x = 0$$

# Phase 2: Two-Dimensional (2D) Kinematics

## 4. Trajectories & Projectile Motion (Syllabus Sections 1–2)

*For vector motion ($r$). You split the problem cleanly into independent horizontal and vertical axes.*

### Horizontal Axis ($a_x = 0$)

* $$x_f = x_i + v_{0x}t$$


* $$v_x = v_{0x} = v_0\cos\theta$$



### Vertical Axis ($a_y = -g$)

* $$y_f = y_i + v_{0y}t - \frac{1}{2}gt^2$$


* $$v_y = v_{0y} - gt$$


* $$v_y^2 = v_{0y}^2 - 2g\Delta y$$


* $$v_{0y} = v_0\sin\theta$$



---

## 5. Circular Motion (Syllabus Section 3)

***[NEW SECTION]** Required for the final week of 2D kinematics. Use these when an object moves in a circular loop.*

* **Centripetal / Radial Acceleration ($a_c$ or $a_r$):** *(Points toward the center of the loop)*

$$a_c = \frac{v^2}{r}$$


* **Tangential Velocity ($v$):** *(Linear speed along the edge)*

$$v = \frac{2\pi r}{T} \quad \text{(where } T \text{ is the Time Period for 1 full loop)}$$


* **Angular Velocity ($\omega$) to Translational Connection:**

$$v = \omega r$$



---

## 6. Relative Motion & Galilean Transformations (Syllabus Section 4)

***[NEW SECTION]** Required for the final 2D homework. Used for measuring objects from different moving frames of reference (e.g., a butterfly's speed as seen by a walking hunter vs. a static camera).*

* **Relative Velocity Vector Addition:**

$$\vec{v}_{\text{Object/Frame A}} = \vec{v}_{\text{Object/Frame B}} + \vec{v}_{\text{Frame B/Frame A}}$$



---

## 7. Non-Constant Acceleration Calculus (The Final Boss Backup)

*Keep this at the very bottom. You won't need it for the core algebraic homework packages, but it serves as your safety net for any complex derivative/integral concept questions.*

* **Down the Ladder (Derivatives):** $a(t) = \frac{dv}{dt} = \frac{d^2x}{dt^2}$
* **Up the Ladder (Integrals):** $v(t) = v_0 + \int_{0}^{t} a(t) \, dt \quad \mid \quad x(t) = x_i + \int_{0}^{t} v(t) \, dt$
