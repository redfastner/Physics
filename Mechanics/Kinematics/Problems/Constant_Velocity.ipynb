### Question 1 (Solve for Final Position, $x_f$)

A cyclist starts their workout at a spatial launchpad anchor of $x_i = -25\text{ meters}$ (meaning they are standing 25 meters behind the designated zero-marker on the track).

They click their stopwatch and begin riding at a perfectly constant velocity of $v = 8\text{ m/s}$ in the positive direction. They maintain this steady cruise for exactly $t = 12\text{ seconds}$.

What is their final position ($x_f$) on the map when the clock hits 12 seconds?

Given:  

$x_i = -25m, v = 8m/s, t = 12s$  

Equation:  

$x_f = x_i + vt$  

Solve:

$x_f = -25 + 8 * 12 = 71m$

```python
print(-25 + 8 * 12)
```

### Question 2 (Solve for Initial Position, $x_i$)

A remote-controlled drone is flying backward along a straight line at a perfectly uniform, steady velocity of $v = -5\text{ m/s}$ (the negative sign means it is moving left/backward on the map).

The operator runs the drone at this steady speed for exactly $t = 15\text{ seconds}$. At the end of this sprint, the drone's final location is measured at the anchor point $x_f = 10\text{ meters}$.

Where on the map did the drone start ($x_i$) when the operator first clicked the stopwatch?

Given:

v = -5m/s, t = 15s, x_f = 10m

Equation:
xi = xf - vt  

Solve:
xi = 10 - (-5 * 15) = 85m

```python
print(10 - (-5 * 15))
```

### Question 3 (Solve for Velocity, $v$)

A automated rover on a factory floor starts its journey at an initial launchpad anchor of $x_i = 42\text{ meters}$.

It moves in a straight line at a completely uniform, steady speed. When the clock hits exactly $t = 8\text{ seconds}$, its final location on the map is recorded at $x_f = 14\text{ meters}$.

Calculate the rover's constant velocity ($v$). Pay close attention to the direction of its motion!

Given:  

xi = 42m, t = 8s, xf = 14m

Equation:  

v avg = delta x / delta t 

Solve:

v avg = (14 - 42) / 8 = 8.75m

```python
print((14 - 42) / 8)
```
