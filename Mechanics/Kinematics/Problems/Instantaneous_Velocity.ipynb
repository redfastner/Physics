You are looking at a **Position-vs-Time graph** of a moving cart.

* From $t = 0\text{ s}$ to $t = 3\text{ s}$, the graph is a perfectly flat, horizontal line sitting at the $x = 5\text{-meter}$ mark.
* At exactly $t = 2\text{ seconds}$, what is the cart's **instantaneous velocity**?

Explain your common-sense reasoning using the slope framework.

At $t = 2\text{ seconds}$, the cart is at $x = 5\text{ m}$.

A tiny fraction of a second later, the cart is still at $x = 5\text{ m}$ (because the line is perfectly flat from $0$ to $3$ seconds).

So, the change in position ($\Delta x$) is 0
