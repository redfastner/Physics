### Question 1 (The Total Mix: Drop Zone)

A construction worker accidentally knocks a wrench off the edge of a high-rise building scaffolding. The wrench starts completely from rest ($v_0 = 0\text{ m/s}$).

It falls straight down under the pure influence of gravity ($a = -9.8\text{ m/s}^2$). A bystander on the ground clicks a stopwatch and notes that it takes exactly $t = 3.5\text{ seconds}$ for the wrench to slam into the dirt.

What is the total **vertical displacement ($\Delta y$)** of the wrench during its fall?

*(Hint: Treat this exactly like your horizontal acceleration problems, but swap out $a$ for gravity!)*

Given:  

v0 = 0m/s, a = -9.8m/s^2, t = 3.5 -> 0  

Eq:  

xf = xi + v0t + 1/2at^2  

Solve:  

delta y = 0 + 0 * 3.5 + -9.8 * 1/2 * 3.5^2 = -60.035m

### Question 2

Given:  

constant speed = 2.3m/s from a -> b, speed = 1.1m/s from b -> a, find avg speed  

Eq:  

avg_speed = total_distance / total_time  

Solve:  

Speed v1 = 2.3m/s, Distance = d, t1 = d/2.3  
Speed v2 = 1.1m/s, Distance = d, t2 = d/1.1  

total_d = d + d = 2d  

total_t = t1 + t2 D/2,3 + D/1.1  

avg_speed = 2D / (D/2.3 + D/1.1)  

d in every term cancel out  

avg_speed = 2 / (1/2.3 + 1/1.1)  

1/2.3 approx= 0.4348, 1/1.1 approx= 0.9091  

0.4348 + 0.9091 = 1.3439  

avg_speed = 2 / 1.3439 approx= 1.49m/s

### Question 7  

5m tall elevator travels at 8m/s up, bolt falls from top, ignore air, how many s till bolt passes bottom of elevator?  

Given:  

vi = 0, delta_y = -5m, a = 9.8m/s^2  

Solve:  

delta_y = v0t + 1/2at^2  

-5 = 0 * t + -4.8t^2  

-5 = -4.8t^2  

t^2 = -5 / -4.9 approx= 1.0204s -> sqrt= 1.01s
