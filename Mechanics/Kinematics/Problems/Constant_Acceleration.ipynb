### Question 1 (Solve for Final Velocity, $v$)

A sports car is idling at a red light, completely stopped ($v_0 = 0\text{ m/s}$). The light turns green, and the driver steps on the gas, creating a constant acceleration of $a = 6\text{ m/s}^2$.

The driver holds the pedal down for a duration of $t = 4\text{ seconds}$.

What is the car's **final velocity ($v$)** at the 4-second mark? Show your givens and pick your equation engine!

Given:  

v0 = 0m/s, a = 6m/s^2, t = 4s  

Eq:  

v = v0 + at  

Solve:  

v = 0 + 6 * 4 = 24m/s  

### Question 2 (Solve for Distance, $\Delta x$)

An airplane sits at the very beginning of a runway, completely at rest ($v_0 = 0\text{ m/s}$). The pilot guns the engines, providing a steady constant acceleration of $a = 4\text{ m/s}^2$ down the straight tarmac.

The plane storms down the runway for exactly $t = 15\text{ seconds}$ before its wheels lift off the ground.

What is the total **distance ($\Delta x$)** the plane traveled down the runway during its takeoff roll?
*(Pick the equation that builds distance using the clock!)*

Given:  

v0 = 0m/s, a = 4m/s^2, t = 15s  

Eq:  

xf = xi + v0t + 1/2at^2

Solve:  

v0t = 0 * 15 = 0  

1/2*a*t^2 = 1/2 * 4 * (15)^2  

15^2 = 225, 4 * 225 = 900  

xf = xi (0) + 1/2 * 900 = 450m

### Question 3 (The Timeless Sprint)

A bullet is fired down a straight gun barrel that is exactly $\Delta x = 0.8\text{ meters}$ long. The bullet starts completely at rest ($v_0 = 0\text{ m/s}$) inside the chamber.

As the gunpowder explodes, it drives the bullet forward with a massive, constant acceleration of $a = 250,000\text{ m/s}^2$.

What is the bullet's **final velocity ($v$)** the exact millisecond it exits the end of the barrel?
*(Hint: Look closely at your equation list for the engine that completely ignores time ($t$))!*

Given:  

delta x = 0.8m (length of barrel), v0 = 0m/s, a = 250,000m/s^2  

Eq:  

v = sqrt{v0^2 + 2a * delta x}

Solve:  

v = sqrt{0 + 250,000 * 2 * 0.8} = 632m/s
